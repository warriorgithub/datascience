```python
##variance ploting before outliar

sns.distplot(funding_data['raised_amount_usd'])
```

```python
acutal = []
def find_outliers(funding):
    population_mean = funding.mean()
    population_std = funding.std()
    anamoly_cut_off = population_std*3
    
    lower_limit = population_mean - anamoly_cut_off
    upper_limit = population_mean + anamoly_cut_off 
    print(lower_limit**2)
    print(upper_limit)
    for outlier in funding:
        if outlier < upper_limit and outlier > lower_limit:
            acutal.append(outlier)
    return acutal;
```

```python
fuding_with_no_outliers = find_outliers(funding_data['raised_amount_usd'])
len(fuding_with_no_outliers)
```

```python
funding_data.head()
amt = funding_data[funding_data['raised_amount_usd'].isin(fuding_with_no_outliers)];
filter_funds = pd.DataFrame(amt)
```

## One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dummy_le = dummy_funds
dummy_le.head(2)
```

```python

dummy_le.company_market = le.fit_transform(dummy_le.company_market)
dummy_le.shape
```

```python
X = dummy_le[['company_market','funded_year']].values
X
```

```python
y = dummy_le.raised_amount_usd
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
ohe = OneHotEncoder(categorical_features=[0])  
```

```python
X = ohe.fit_transform(X).toarray()
```

```python
X
```

```python
olm = LinearRegression()
olm.fit(X,y)
```

```python
o_predict_funding = olm.predict(X)
o_predict_funding
```

## Do Company Domain play a major role in funding ?

```python
olm.score(X,y)
```
