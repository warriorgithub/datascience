```python
##variance ploting before outliar

sns.distplot(funding_data['raised_amount_usd'])
```

```python
acutal = []
def find_outliers(funding):
    population_mean = funding.mean()
    population_std = funding.std()
    anamoly_cut_off = population_std*3
    
    lower_limit = population_mean - anamoly_cut_off
    upper_limit = population_mean + anamoly_cut_off 
    print(lower_limit**2)
    print(upper_limit)
    for outlier in funding:
        if outlier < upper_limit and outlier > lower_limit:
            acutal.append(outlier)
    return acutal;
```

```python
fuding_with_no_outliers = find_outliers(funding_data['raised_amount_usd'])
len(fuding_with_no_outliers)
```

```python
##variance ploting after outliar
## It seems we have skew here towards right
sns.distplot(fuding_with_no_outliers,color='b')
```

```python
##To see more precise variance I delibrately ristricted by data point less then < 20000000 

df_filter = funding_data.query('raised_amount_usd<20000000')
df_filter.describe()
```

```python
sns.distplot(df_filter['raised_amount_usd'],color='b')
```

```python
#population_mean = np.mean(fuding_with_no_outliers)
mean = df_filter['raised_amount_usd'].mean() 
mean
```

```python
funding_data.columns
```

```python
df_filter[['company_name','funded_at']] = df_filter[['company_name','funded_at']].astype(int)
df_filter[['raised_amount_usd']] = df_filter[['raised_amount_usd']].astype(float)
```

```python

categ_fetures = df_filter.dtypes==int
cate_cols = df_filter.columns[categ_fetures].to_list()
cate_cols
```

## Label Encoder 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df_filter[cate_cols] = df_filter[cate_cols].apply(lambda col: le.fit_transform(col))
```

```python
rem_data = df_filter.drop(df_filter[cate_cols],axis=1)
rem_data.head()
```

```python
ecoded_feature = df_filter[cate_cols]
```

```python
fin_fund = pd.concat([ecoded_feature,rem_data],axis=1)
fin_fund.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = fin_fund.drop(['raised_amount_usd'],axis=1)
```

```python
lm = LinearRegression()
lm
```

```python
y = fin_fund.raised_amount_usd
```

```python
lm.fit(X,y)
```

```python
pd.DataFrame({'feature': X.columns, 'estimateCofficients':lm.coef_})
```

```python
predict_funding = lm.predict(X)
predict_funding
```

```python
lm.score(X,y)
```

```python

plt.hist(predict_funding)
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.scatter(predict_funding,y)
plt.xlabel('Predicted Funding')
plt.ylabel('Actual Funding')
plt.show()
```
