```python
import numpy as np
import pandas as pd
```

```python
missing = np.nan      #set nan vlaue to missing variable (Nan is not a number)
```

```python
series_obj = pd.Series(['row1','row2',missing,'row4',missing])
```

```python
series_obj
```

### Indentify missing values

```python
series_obj.isnull()
```

### Fill missing values

```python
np.random.seed(25)     #replicate same random number everytime
```

```python
df = pd.DataFrame(np.random.rand(36).reshape(6,6))
df
```

```python
#Assign missing value to dataframe
df.loc[3:5,0]=missing
df.loc[1:4,5]=missing
df
```

```python
df.fillna(0)
```

**fill values with dictonary**

```python
df.fillna({0:0.1,5:1.25})
```

**fill using fill foward / backword**

**fill using last or first know value in columns**

```python
df.fillna(method='ffill')
```

### Couting missing values

```python
df.isnull().sum()
```

### Filtering out missing values

```python
df.dropna(axis=1)   #drop column wise null values
```

```python
df.dropna(how='all')    #drop missing value row wise(row has all missing values)
```
