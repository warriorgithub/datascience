```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.io.stata.read_stata('data/us_job_market_discrimination.dta')
```

```python
w = data[data.race=='w']
b = data[data.race=='b']
```

```python
# number of callbacks for black-sounding names
w_call = sum(data[data.race=='w'].call)
w_n = len(w)
w_r = (w_call / w_n)
```

```python
w_size = data[data.race=='w'].race.size
w_size
```

```python
b_call = sum(data[data.race=='b'].call)
b_n = len(b)
b_r = (b_call / b_n)
```

```python
b_size = data[data.race=='b'].race.size
b_size
```

```python
w_call_percentage = (w_call/w_size)
b_call_percentage = (b_call/b_size)
print("Percentage of white people got the call",w_call_percentage)
print("Percentage of black people got the call",b_call_percentage)
print("differe between  ",w_call_percentage - b_call_percentage)
```

**Q2.What are the null and alternate hypotheses**

H0 :  The rate at which white-sounding names and black-sounding names receive callbacks is equal

**We have sample more then 30 so we can perform z test**

```python
w_sample = np.random.binomial(w_size,w_call_percentage,1000)
b_sample = np.random.binomial(b_size,b_call_percentage,1000)
#wx, wy = ecdf(w_sample)
_ = plt.plot(w_sample, marker = '.', linestyle = 'none')

#bx, by = ecdf(b_sample)
_ = plt.plot(b_sample, marker = '.', linestyle = 'none')

# Time to label the plot
sns.set(rc={'figure.figsize':(12,8)})
_ = plt.xlabel('Quantity of Callbacks')
_ = plt.ylabel('CDF')
_ = plt.title('Black-Sounding vs. White-Sounding Names')
_ = plt.legend(('White-Sounding Names', 'Black-Sounding Names'), loc = 'lower right')
```

**Q3. Compute margin of error, confidence interval, and p-value. Try using both the bootstrapping and the frequentist statistical approaches**

```python
def ztest_proportions_two_samples(w_c, n1, b_c, n2, one_sided = False):
    """This will return the z-score and p-value for a 2-sample Z-test of proportions"""
    p1 = w_c / n1
    p2 = b_c / n2
    #compute p-value
    p = (w_c + b_c) / (n1 + n2)
   
    se = np.sqrt(p * (1 - p) * (1 / n1 + 1 / n2))
    #compute z-statistic
    z = (p1 - p2) / se
    p = 1 - stats.norm.cdf(abs(z))
    p *= 2 - one_sided
    return z, p
```

```python
proportions_diff = w_call_percentage - b_call_percentage

z_score = 1.96 # 95% confidence z-score
p_r1 = w_r * (1 - w_r) / w_n
p_r2 = b_r * (1 - b_r) / b_n

ci_low = proportions_diff - z_score * (np.sqrt(p_r1 + p_r2))
ci_high = proportions_diff + z_score * (np.sqrt(p_r1 + p_r2))

z_val, p_val = ztest_proportions_two_samples(w_call, w_n, b_call, b_n)

moe = (ci_high - ci_low) / 2

# Print Margin of error, confidence interval values, and p-value.
print('The Margin of Error is: +/-{:.5}'.format(moe))
print('A 95% Confidence Interval: {:.5} - {:.5}'.format(ci_low, ci_high))
print('The p-value is: {:.5}'.format(p_val))
print('We observed a difference in proportions of: {:.2%}'.format(proportions_diff))
print('The z-value is: {:.5}'.format(z_val))
```

With a 95% confidence interval, it can be concluded that for samples from the same population, the difference in proportions will be less than the empircal results
